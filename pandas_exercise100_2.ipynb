# 57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
Unknown : N

Less than $\$40K$ : a

$\$40K - \$60K$ : b

$\$60K - \$80K$ : c

$\$80K - \$120K$ : d

$\$120K +'$ : e

```python
df['newIncome']=df.Income_Category.map({'Unknown':'N','Less than $40K':'a','$40K - $60K':'b','$60K - $80K':'c','$80K - $120K':'d','$120K +':'e'})
df['newIncome'].head()
```

# 58.Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라 Unknown : N
Unknown : N

Less than $\$40K$ : a

$\$40K - \$60K$ : b

$\$60K - \$80K$ : c

$\$80K - \$120K$ : d

$\$120K +'$ : e

```python
def change(df):
  if df == 'Unknown':
    return 'N'
  elif df =='Less than $40K':
    return 'a'
  elif df== '$40K - $60K':
    return 'b'
  elif df =='$60K - $80K':
    return 'c'
  elif df =='$80K - $120K':
    return 'd'
  elif df =='$120K+':
    return 'e'

df['newIncome'] = df.Income_Category.apply(change)
df['newIncome'].head()
```

```python
def change_year(df):
  import datetime

  year = df.year - 100 if df.year >=2061 else df.year
  return pd.to_datetime(datetime.date(year,df.month,df.day))
```

# 76. 년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
df = pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
```

```python
def change_date(df):
  import datetime

  hour = df.split(':')[1]
  date = df.split(':')[0]

  if hour =='24':
    hour='00:00:00'

    Finaldate = pd.to_datetime(date+" "+hour) + datetime.timedelta(days=1)
  
  else:
    hour = hour +":00:00"
    Finaldate = pd.to_datetime(date+ " "+ hour)

  return Finaldate
```

```python
df['(년-월-일:시)'] = df['(년-월-일:시)'].apply(change_date)
df.head()
```
