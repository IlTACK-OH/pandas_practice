## 23. quantity, item_price 두 개의 컬럼으로 구성된 새로운 데이터프레임을 정의하라

## 32. df의 item_name컬럼 값이 Steak Salad 또는 Bowl인 데이터를 인덱싱하라.

```python
df_isin = df[df['item_name'].isin(['Steak Salad','Bowl'])]
display(df_isin)

# 인덱싱하라? -> loc나 iloc가 인덱싱하는 방법이라고 한다.
df_loc = df.loc[(df.item_name =='Steak Salad')|(df.item_name=='Bowl')]
display(df_loc)
```

## 33. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

```python
df_isin[df_isin.duplicated('item_name') == False]
```

```python
# 사이트 정답 예시 코드
df.loc[(df.item_name =='Steak Salad') | (df.item_name =='Bowl')].drop_duplicates('item_name')
```

## 34. df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 마지막 케이스만 남겨라

```python
df.loc[(df.item_name =='Steak Salad') | (df.item_name =='Bowl')].drop_duplicates('item_name',keep='last')
```

## 35. df의 데이터 중 new_price값이 new_price값의 평균값 이상을 가지는 데이터들을 인덱싱하라
