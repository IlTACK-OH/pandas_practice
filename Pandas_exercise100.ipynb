# 20. 데이터를 로드하라

```python
df =  pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/chipo.csv')
df.head()
```

# 21. quantity컬럼 값이 3인 데이터를 추출하여 첫 5행을 출력하라.

```python
df[df['quantity']==3].head()
```

# 22. quantity컬럼 값이 3인 데이터를 추출하여 index를 0부터 정렬하고 첫 5행을 출력하라

```python
df[df['quantity']==3].reset_index(drop=True).head(5)
```

# 23. quantity, item_price 두 개의 컬럼으로 구성된 새로운 데이터프레임을 정의하라

```python
df_new = df[['quantity','item_price']].reset_index(drop=True).copy()
df_new.head()
```

# 24. Item_price 컬럼의 달러표시 문자를 제거하고, float타입으로 저장하여 new_price컬럼에 저장하라

```python
df['new_price'] = df['item_price'].str.replace('$','',regex=True).astype('float')
df['new_price'].head()
```

# 25. new_price컬럼이 5이하의 값을 가지는 데이터프레임을 추출하고, 전체 갯수를 구하여라

```python
df[df['new_price'] <= 5].shape[0]
```

# 26. item_name명이 Chicken Salad Bowl인 데이터 프레임을 추출하라고 index값을 초기화 하여라

```python
df[df['item_name']=='Chicken Salad Bowl'].reset_index(drop=True).head(3)
```

# 27. new_price값이 9이하이고 item_name값이 Chicken Salad Bowl인 데이터 프레임을 추출하라

```python
df[(df['new_price']<=9) & (df['item_name']=='Chicken Salad Bowl')].head(5)
```

# 28. df의 new_price 컬럼 값에 따라 오름차순으로 정리하고  index를 초기화 하여라

```python
df.sort_values(by = 'new_price', ascending = True).reset_index(drop=True).head()
```

# 29. df의 item_name 컬럼 값 중 Chips 포함하는 경우의 데이터를 출력하라.

```python
df[df['item_name']=='Chips'].head()
```

이러면 틀리다. 포함하는 경우라 하였으므로 다음과 같이 코드를 작성해야 한다.
