# 앤스콤 데이터 집합 불러오기

데이터 시각화를 보여주는 전형적인 사례로 엔스콤 4분할 그래프(Anscombe's quarter)가 있다. 이 그래프는 영국의 프랭크 엔스콤이 데이터를 시각화하지 않고 수치만 확인할 때 발생할 수 있는 함정을 보여주기 위해 만든 그래프이다.
<br><br>
엔스콤 4분할 그래프의 모든 데이터 그룹은 x,y열을 가지고 있다. 그런데 이 4개의 데이터 그룹은 각각 평균, 분산과 같은 수칫값이나 상관관계, 회귀선이 같다는 특징이 있다. 이러한 특성 때문에 **"4개의 데이터 그룹은 모두 같을 것이다"**라는 착각을 일으킨다. 이것이 엔스콤이 지적한 `함정`이다.<br><br>
이런 점에서 데이터 시각화는 데이터 분석 분야에서 아주 중요한 요소라고 할 수 있다.

---
엔스콤 데이터 집합은 `seaborn`라이브러리에 포함되어 있다.`load_dataset`메서드에 문자열 `anscombe`을 전달하면 엔스콤 데이터를 불러올 수 있다. 이때 데이터프레임의 열 이름 중 dataset 열이 데이터 그룹을 구분한다는 것을 알 수 있다.

```python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
display(anscombe)
print(type(anscombe))
```

# 한 번에 4개의 그래프 그리기

## matplotlib.pyplot 라이브러리로 그래프 그리기
1. 전체 그래프가 위치할 기본 틀을 만든다.
2. 그래프를 그려 넣을 그래프 격자를 만든다.
3. 그런 다음 격자에 그래프를 하나씩 추가한다. 격자에 그래프가 추가되는 순서는 왼쪽에서 오른쪽 방향이다.
4. 만약 격자의 첫 번째 행이 꽉 차면 두 번째 행에 그래프를 그려 넣는다.

## 그래프 출력하기

```python
# 각 data 저장
dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']

# 그래프 격자가 위치할 기본 틀을 만든다.
fig = plt.figure()

# add_subplot 메서드로 그래프 격자 그리기
axes1 = fig.add_subplot(2,2,1)
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)
```

이제 plot 메서드에 데이터를 전달하여 그래프를 출력한다.

```python
axes1.plot(dataset_1['x'],dataset_1['y'],'o')
axes2.plot(dataset_2['x'],dataset_2['y'],'o')
axes3.plot(dataset_3['x'],dataset_3['y'],'o')
axes4.plot(dataset_4['x'],dataset_4['y'],'o')

fig
```

```python
# 그래프 격자에 제목 추가
axes1.set_title("dataset_1")
axes2.set_title("dataset_2")
axes3.set_title("dataset_3")
axes4.set_title("dataset_4")

fig
```

## 다변량 그래프

---
앞에서 실습한 산점도 그래프에 성별을 새 변수로 추가해 보겠다.

```python
# 함수 정의
def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

seaborn라이브러리를 잘 활용하면 matplotlib보다 좀 더 화려한 그래프를 그릴 수 있다.

---
다시 tips데이터를 불러온 후 seaborn라이브러리로 히스토그램을 출력하여 보겠다.<br><br>
`subplots`,`distplot`메서드를 사용하면 seaborn라이브러리로 히스토그램을 출력할 수 있다.<br><br>
먼저 `sunplots`메서드로 기본 틀을 만들고 `distplot`메서드에 `total_bill` 열 데이터를 전달하면 히스토그램을 그릴 수 있다.<br><br>
이때, `distplot`메서드를 사용하면 히스토그램과 밀집도 그래프를 같이 그리는데 만약 밀집도 그래프를 제외하고 싶을 경우 `kde`인잣값을 `False로 설정한다.

---
책과 달리 seaborn이 업데이트 되어 distplot말고 histplot이나 displot을 사용하라는 경고 메세지가 나타난다.

```python
# kde = True
tips = sns.load_dataset('tips')
# displot을 이용한 출력
ax = sns.displot(tips['total_bill'], kde = True)
# histplot을 이용한 출력
ax = plt.subplots()
ax = sns.histplot(tips['total_bill'], kde = True)
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
# kde = False
tips = sns.load_dataset('tips')
# displot을 이용한 출력
ax = sns.displot(tips['total_bill'], kde = False)
# histplot을 이용한 출력
ax = plt.subplots()
ax = sns.histplot(tips['total_bill'], kde = False)
ax.set_title('Total Bill Histogram with Density Plot')
```

만일 밀집도 그래프만 남겨두고 싶다면 `hist`인자를 False로 지정하면 된다.

---
버전이 바뀐 것인지 내가 못찾는 것인지 일단 밀집도만 출력하기 위해서는 kdeplot을 사용한다.

>참고사이트
https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot

```python
ax = plt.subplots()
ax = sns.kdeplot(tips['total_bill'])
ax.set_title('Total Bill Density')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

히스토그램과 밀집도를 그리는 메서드에 rug인자를 추가하여 양탄자그래프까지 그려보도록 하겠다.

```python
ax = plt.subplots()

ax = sns.histplot(tips['total_bill'],kde =True,color ='g')
ax = sns.rugplot(tips['total_bill'],expand_margins = False)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

이번에는 히스토그램과 비슷하게 생긴 Count 그래프를 그려보겠다. `count`그래프는 이산값을 나타낸 그래프이다.

## 이변량 그래프 그리기
이변량 그래프 중 산점도를 그려보도록하겠다.<br>
이번 실습에서는 `regplot`메서드를 사용하여 본다.<br>
`regplot`메서드를 사용하면 산점도 그래프와 회귀선을 함께 그릴 수 있다.<br>
 만약 회귀선을 제거하고자 하면 fit_reg인자를 False로 지정하면 된다.

```python
# 회귀선과 함께 출력
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill',y='tip', data = tips)
ax.set_title('Scatterplot of Totla Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')

# 회귀선을 제거
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill',y='tip', data = tips, fit_reg = False)
ax.set_title('Scatterplot of Totla Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

산점도 그래프와 히스토그램을 한 번에 그려주는 `jointplot`메서드를 사용하여보겠다.<br><br>
`x`,`y`인자에 원하는 열 이름을 지정하고 data인잣값으로 데이터프레임을 지정한다.

```python
joint = sns.jointplot(x='total_bill', y='tip',data= tips)
joint.set_axis_labels(xlabel ='Total Bill',ylabel = 'Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip',fontsize = 10, y =1.03)
```

산점도가 겹쳐 확인하기 어렵다는 단점이 존재한다.<br>이때 육각그래프를 사용하면 이러한 단점을 해결할 수 있다.<br><br>
육각그래프는 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프이다. 그래서 특정 데이터의 개수가 많아지면 점점 진한색으로 표현된다.<br><br>
육각 그래프는 `jointplot`메서드를 그대로 사용한다. 대신, `kind`인잣값을 `hex`로 지정하면 된다.

```python
hexbin = sns.jointplot(x='total_bill', y='tip',data= tips, kind = 'hex')
hexbin.set_axis_labels(xlabel ='Total Bill',ylabel = 'Tip')
hexbin.fig.suptitle('Joint Plot of Total Bill and Tip',fontsize = 10, y =1.03)
```

## 이차원 밀집도
이차원 밀집도는 kdeplot메서드로 만들 수 있다. 다음은 total_bill,tip 열 데이터를 전달하여 이차원 밀집도를 그린 것이다. 이때, shade 인잣값을 True로 지정하면 그래프에 음영 효과를 줄 수 있다.

```python
ax = plt.subplots()
ax = sns.kdeplot(x=tips['total_bill'],
                 y=tips['tip'],
                 shade = True)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total bill')
ax.set_ylabel('Tip')
```

## 박스 그래프 그리기
박스 그래프는 최솟값, 1분위수, 중간값, 3분위수, 최댓값, 이상치 등 다양한 통계량을 한 번에 표현하는 그래프이다.

```python
ax = plt.subplots()
ax = sns.boxplot(x='time',y='total_bill',data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

박스 그래프는 다양한 통계 수치를 확인하기 위해 자주 사용하는 그래프지만 데이터 분산이 모호하게 표현된다.<br> 이런 경우에는 박스 그래프에 커널 밀도를 추정한 바이올린 그래프를 사용하면 된다.<br> 쉽게 말해 박스형태를 바이올린 형태로 바꾸면 바이올린 그래프가 된다.<br>바이올린 그래프는 `violinplot`메서드를 사용하여 그릴 수 있다.

```python
ax = plt.subplots()
ax = sns.violinplot(x='time',y='total_bill',data=tips)
ax.set_title('Violin plot of total bill by time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Total Bill')
```

## 관계 그래프 그리기
관계 그래프는 지금까지 설명한 그래프를 종합한 그래프라고 생각하면 된다. 관계 그래프는 `pairplot`메서드에 데이터프레임을 넣는 방법으로 간단하게 그릴 수 있다.

```python
fig = sns.pairplot(tips)
```

이때 관계 그래프는 중복된 정보가 표현된다는 단점이 있다. 즉, 관계그래프의 절반(대각선 기준)은 나머지 절반과 동일하다.<br>이런 경우에는 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체하면 된다.

```python
pair_grid = sns.PairGrid(tips)
pair_gird = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot,rug=True)
plt.show()
```

# 데이터프레임과 시리즈로 그래프 그리기

데이터프레임과 시리즈로도 그래프를 그릴 수 있다.<br> 간편하게 그래프를 그리고 싶은 경우에는 데이터프레임이나 시리즈를 사용하여 바로 그래프를 그릴 수 있다.<br>

---
먼저 `plot`속성에 정의된 `hist`메서드를 사용하면 해당 시리즈의 값을 이용하여 히스토그램을 바로 그릴 수 있다.

```python
ax = plt.subplots()
ax = tips['total_bill'].plot.hist()
```

투명도를 조절하려면 `hist`메서드의 `alpha`,`bins`,`ax`인자를 사용하면 된다.<br><br>
`alpha`: 투명도
`bins`: x축의 간격

```python
fig, ax = plt.subplots()
ax = tips[['total_bill','tip']].plot.hist(alpha=0.5,bins=20,ax=ax)
```

밀집도, 산점도 그래프, 육각 그래프는 각각 kde,scatter,hexbin메서드를 사용하여 그릴 수 있다.

```python
ax = plt.subplots()
ax = tips['tip'].plot.kde()
```

```python
fig,ax = plt.subplots()
ax = tips.plot.scatter(x='total_bill',y='tip',ax=ax)
```

```python
fig,ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill',y='tip',ax=ax)
```

육각형 그래프의 경우 `gridsize` 인자를 사용하여 변경 가능하다.

```python
fig,ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill',y='tip',gridsize=10,ax=ax)
```

```python
fig,ax = plt.subplots()
ax = tips.plot.box(ax=ax)
```

# 그래프에 스타일 적용하기


```python
fig,ax = plt.subplots()
ax = sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

```python
sns.set_style('whitegrid')
fig,ax = plt.subplots()
ax = sns.violinplot(x='time',y='total_bill',hue='sex',data=tips,split=True)
```

```python
fig = plt.figure()

seaborn_styles = ['darkgrid','whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position= idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin = sns.violinplot(x='time',y='total_bill',data=tips,ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
